## Average Revenue by category 

```python
sorted_df = df.groupby('category')['revenue_usd'].mean().sort_values(ascending=False).index

sns.barplot(
    data=df,
    x='category',
    y="revenue_usd",
    hue='traffic_source',
    order=sorted_df
)
plt.title('Average Revenue by category')
plt.xlabel('Category')
plt.ylabel('Revenue ($)')
```

1. The average revenue for the `Electronics` category is the highest, while `Books` have the lowest revenue.
2. For `Furniture`, the revenue is exclusively represented by Google Ads. This may indicate that other sources were not utilized or data for them is unavailable.
3. For `Electronics` and `Clothing`, the highest revenue comes from Organic Search. For `Books`, the revenue from all sources is low and nearly equal, which may suggest weak marketing efforts or low demand for the category.
4. The absence of Google Ads in other categories might be due to a lack of investment in this channel or incomplete data.

##  Units Sold and Revenue

```python
sns.scatterplot(data, x='units_sold', y='revenue_usd', s=100, hue='category')
```

We see linear relationships between the number of units sold and revenue for every category. This is expected, as the revenue is calculated as the product of the number of units sold and the price per unit. Also we see that profit from `Electronics` and `Furniture` is higher than from `Books` and `Clothing`, because the price per unit is higher for the former categories.
